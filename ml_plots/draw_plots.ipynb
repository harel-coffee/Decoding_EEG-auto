```python
def export_legend(legend, filename="exp_legend_ieee", expand=[-5,-5,5,5]):
    fig  = legend.figure
    fig.canvas.draw()
    bbox  = legend.get_window_extent()
    bbox = bbox.from_extents(*(bbox.extents + np.array(expand)))
    bbox = bbox.transformed(fig.dpi_scale_trans.inverted())
    fig.savefig("/gdrive/MyDrive/Project_DEAP/4.1.2021/plots_ieee/" + filename + '.svg', dpi="figure", bbox_inches=bbox)
```

```python
def va_features_plots(dataset_list):
    legendList = [];
    for idx,dataset in enumerate(dataset_list):

        os.chdir("/gdrive/MyDrive/Project_DEAP/4.1.2021/{}/plots/".format(dataset))

        df = pd.read_csv("topCommonFeatureFSRegressionRankingSelectKBest11.csv")
        # plt.plot(df['RMSE'])
        fig = plt.gcf()
        fig.set_size_inches(16, 10)
        # X = pd.DataFrame([x for x in range(1,) ])
        plt.rcParams.update({'font.size': 40})
        plt.xlabel('Top N Features')
        plt.ylabel('RMSE') 
        # df.index += 1
        # X = pd.DataFrame([str(x) for x in df.index])
        plt.xticks(np.arange(0,25,5),np.arange(1,26,5))
        plt.plot(df.loc[:,"RMSE"], linewidth=2.0, markersize = 10)
        
        plt.tight_layout()
        # print(pd.DataFrame(df.index))
        legendList.append(f'{dataset}(Valence)');

        os.chdir("/gdrive/MyDrive/Project_DEAP/4.1.2021/{}/arousal_plots/".format(dataset))


        df = pd.read_csv("topCommonFeatureFSRegressionRankingSelectKBest11.csv")
        # plt.plot(df['RMSE'])
        fig = plt.gcf()
        fig.set_size_inches(16, 10)
        # X = pd.DataFrame([x for x in range(1,) ])
        plt.rcParams.update({'font.size': 40})
        plt.xlabel('Top N Features')
        plt.ylabel('RMSE')
        # df.index += 1
        # X = pd.DataFrame([str(x) for x in df.index])
        plt.xticks(np.arange(0,25,5),np.arange(1,26,5))
        plt.plot(df.loc[:,"RMSE"], '-o', color = f'C{idx}', linewidth=2.0, markersize = 10)
        plt.tight_layout()
        legendList.append(f'{dataset}(Arousal)');
        

    # legend = plt.legend(legendList, bbox_to_anchor=(0.5, -0.6),fontsize='xx-small')
    # legend = plt.legend(legendList,fontsize='xx-small')

    # export_legend(legend)

    ax = plt.gca()
    box = ax.get_position()
    # ax.set_position([box.x0, box.y0 + box.height * 0.1,box.width, box.height * 0.9])

    # Put a legend below current axis
    ax.legend(labels = legendList, loc='upper center', bbox_to_anchor=(0.5, -0.05),fancybox=True, shadow=True, ncol=3)
    
    export_legend(ax.get_legend())

    ax.legend().set_visible(False)
    # ax.legend().set_visible(False)
    plt.savefig("/gdrive/MyDrive/Project_DEAP/4.1.2021/plots_ieee/exp_feature_plot.svg".format(dataset), bbox_inches='tight', dpi=500)

    # plt.savefig("verification/DREAMER_Combined_Features_Plots1.svg", bbox_inches='tight', dpi=500)
    plt.show()
    plt.clf()

        
```

```python
def va_electrodes_plots(dataset_list):
    legendList = []
    for idx, dataset in enumerate(dataset_list):

        os.chdir("/gdrive/MyDrive/Project_DEAP/4.1.2021/{}/plots/".format(dataset))

        df = pd.read_csv("topCommonElectrodeFSRegressionRankingSelectKBest11.csv")
        # plt.plot(df['RMSE'])
        fig = plt.gcf()
        fig.set_size_inches(18,10)
        # X = pd.DataFrame([x for x in range(1,) ])
        plt.rcParams.update({'font.size': 40})
        plt.xlabel('Top N Electrodes')
        plt.ylabel('RMSE')
        # df.index += 1
        # X = pd.DataFrame([str(x) for x in df.index])
        plt.xticks(np.arange(0,14,2),np.arange(1,15,2))
        plt.plot(df.loc[:,"RMSE"], linewidth=2.0, markersize = 10)
        plt.tight_layout()
        legendList.append(f'{dataset}(Valence)');


        os.chdir("/gdrive/MyDrive/Project_DEAP/4.1.2021/{}/arousal_plots/".format(dataset))

        df = pd.read_csv("topCommonElectrodeFSRegressionRankingSelectKBest11.csv")
        # plt.plot(df['RMSE'])
        fig = plt.gcf()
        fig.set_size_inches(18,10)
        # X = pd.DataFrame([x for x in range(1,) ])
        plt.rcParams.update({'font.size': 40})
        plt.xlabel('Top N Electrodes')
        plt.ylabel('RMSE')
        # df.index += 1
        # X = pd.DataFrame([str(x) for x in df.index])
        plt.xticks(np.arange(0,14,2),np.arange(1,15,2))
        plt.plot(df.loc[:,"RMSE"], '-o', color = f'C{idx}', linewidth=2.0, markersize = 10)
        # plt.plot(df.loc[:,"RMSE"], color = 'C1')
        plt.tight_layout()
        legendList.append(f'{dataset}(Arousal)');


    legend = plt.legend(legendList, fontsize='xx-small')
    export_legend(legend)
    ax = plt.gca()
    ax.legend().set_visible(False)
    plt.savefig("/gdrive/MyDrive/Project_DEAP/4.1.2021/plots_ieee/electrode_plot.svg".format(dataset), bbox_inches='tight', dpi=500)
    plt.show()
    plt.clf()
```

```python
def export_legend(legend, filename="legend_new.png", expand=[-5,-5,5,5]):
    fig  = legend.figure
    fig.canvas.draw()
    bbox  = legend.get_window_extent()
    bbox = bbox.from_extents(*(bbox.extents + np.array(expand)))
    bbox = bbox.transformed(fig.dpi_scale_trans.inverted())
    fig.savefig("/gdrive/MyDrive/Project_DEAP/4.1.2021/plots_ieee/" + filename + '.svg', dpi="figure", bbox_inches=bbox)
```
